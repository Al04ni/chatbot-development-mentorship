By starting on the zero shot classification, which help use to classify the text basing on the labels and also give us scores of the label toward the context of the text.


```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification",model="knowledgator/comprehend_it-base")
classifier(
    "This is a course about the Transformers library,and be prepared to do the hard things",
    candidate_labels=["education", "politics", "business","technology","motivation"],
)
```

Shift to text generation pipeline.
adding max_length to exceed the default number of the model,and also num_return_sequences for the number of paragraohs to be generated.

```python
from transformers import pipeline
generate=pipeline("text-generation",
                  max_length=50,#we can also extend basing on how long we want our answers
                  model="openai-community/gpt2",#free quick model we used so far
                  truncation=True,#to allow the model use more words as we want
                  #number of anwers we want
                  num_return_sequences=3)
generate("Let's change our Africa,and be prepared to do the hard things")
```
